```python
import numpy as np
import matplotlib.pyplot as plt

def re_deby(x_array,eps_0,eps_inf,tau):
    return eps_inf+(eps_0-eps_inf)/(1+tau*tau*x_array*x_array)

def im_deby(x_array,eps_0,eps_inf,tau):
    return (eps_0-eps_inf)*tau*x_array/(1+tau*tau*x_array*x_array)
```

```python
x_array=np.arange(1,1000000,10)
# eps_0 = 30000
# eps_inf = 1
# tau = 1000
```

```python
eps_0 = 300
eps_inf = 100
tau = 0.1
plt.xscale("log")
plt.plot(x_array,re_deby(x_array,eps_0,eps_inf,tau),label='real')
plt.plot(x_array,im_deby(x_array,eps_0,eps_inf,tau),label='imag')
plt.xlabel("log (Frequency)")
plt.ylabel("Îµ")
plt.legend()
# plt.ylim(0,50)
# plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.plot()
```

```python
def re_lorentz(x_array,eps_0,eps_inf,tau,omega_0):
    return eps_inf+(eps_0-eps_inf)*(omega_0*omega_0-x_array*x_array)/((x_array*x_array-omega_0*omega_0)**2+tau*tau*x_array*x_array)

def im_lorentz(x_array,eps_0,eps_inf,tau,omega_0):
    return (eps_0-eps_inf)*tau*x_array/((x_array*x_array-omega_0*omega_0)**2+tau*tau*x_array*x_array)
```

```python
omega_0=500
# x_array=np.arange(2*omega_0)
x_array=np.arange(1,10000,10)
tau = 100
eps_0   = 50000
eps_inf = 0.2
plt.xscale("log")
plt.plot(x_array,re_lorentz(x_array,eps_0,eps_inf,tau,omega_0),label='real')
plt.plot(x_array,im_lorentz(x_array,eps_0,eps_inf,tau,omega_0),label='imag')
plt.xlabel("Frequency")
plt.ylabel("Îµ")
# plt.xlim(0,10000)
# plt.ylim(0,1)

plt.legend()
# plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.plot()
```
